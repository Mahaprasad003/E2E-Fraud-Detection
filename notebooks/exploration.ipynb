```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('../data/raw/creditcard.csv')
```

```python
print("Shape of data: ", df.shape)
```

2,84,807 rows and 31 columns.

```python
df.describe()
```

Since max of time is 172792 and min is 0, we can assume time is in seconds. So total time span is around 48 hours.

```python
# check for missing data
total = df.isnull().sum().sort_values(ascending = False)
percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
pd.concat([total, percent], axis=1, keys=['Total', 'Percent']).transpose()
```

There are zero missing values in any column.

```python
# checking fraud/nofraud distributions
print('No Frauds', round(df['Class'].value_counts()[0]/len(df) * 100,2), '% of the dataset')
print('Frauds', round(df['Class'].value_counts()[1]/len(df) * 100,2), '% of the dataset')
```

The dataset is highly unbalanced with the positive class (frauds) accounting for 0.17% of all transactions.

```python
# Visualizing the class distribution vanilla
plt.figure(figsize=(6,4))
df['Class'].value_counts().plot(kind='bar')
plt.title('Class Distribution 0: No Fraud, 1: Fraud')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

Not much to see due to class imbalance. Let's try log scale.

```python
# visualizing the class distribution by taking log of counts
plt.figure(figsize=(6,4))
np.log1p(df['Class'].value_counts()).plot(kind='bar')
plt.title('Class Distribution (Log Scale) 0: No Fraud, 1: Fraud')
plt.xlabel('Class')
plt.ylabel('Log Frequency')
plt.show()
```

We can see that the dataset is highly imbalanced with only a small fraction of fraudulent transactions.

```python
# Visualising amount distirbution
plt.figure(figsize=(10,6))
plt.hist(df['Amount'], bins=50)
plt.title('Transaction Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.yscale('log')
plt.show()
```

This plot shows the distribution of transaction amounts on a logarithmic scale to better visualize the spread of values, especially given the presence of outliers in transaction amounts.

```python
# box plot for amount with classes with showfliers false and true side by side
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.title('Box Plot of Amount by Class (Without Outliers)')
sns.boxplot(x='Class', y='Amount', data=df, showfliers=False)
plt.subplot(1, 2, 2)
plt.title('Box Plot of Amount by Class (With Outliers)')
sns.boxplot(x='Class', y='Amount', data=df, showfliers=True)
plt.show()
```

```python
tmp = df[['Amount','Class']].copy()
class_0 = tmp.loc[tmp['Class'] == 0]['Amount']
class_1 = tmp.loc[tmp['Class'] == 1]['Amount']
class_0.describe()
```

```python
class_1.describe()
```

The fraud transactions have a higher mean but lower maximum amount compared to non-fraud transactions. This suggests that while fraud transactions tend to be of higher value on average, the largest transactions are typically non-fraudulent.

```python
from sklearn.manifold import TSNE
sample_size = 5000
fraud_sample = df[df['Class'] == 1].sample(n=min(492, len(df[df['Class'] == 1])), random_state=42)
non_fraud_sample = df[df['Class'] == 0].sample(n=sample_size - len(fraud_sample), random_state=42)
sample_df = pd.concat([fraud_sample, non_fraud_sample])

X_sample = sample_df.drop(['Class', 'Time'], axis=1)
y_sample = sample_df['Class']

print("Applying t-SNE... this may take a minute...")
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_sample)

plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], 
                     c=y_sample, 
                     cmap='coolwarm', 
                     alpha=0.6,
                     s=50)
plt.colorbar(scatter, label='Class (0=Normal, 1=Fraud)')
plt.title('t-SNE Visualization of Credit Card Transactions')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.grid(True, alpha=0.3)
plt.savefig('../assets/tsne.png', dpi=150, bbox_inches='tight')
plt.show()


print(f"Sample contains {len(fraud_sample)} fraud and {len(non_fraud_sample)} non-fraud transactions")
```

Cool looking pattern, but we can see that the fraud points are quite scattered compared to the dense cluster of non-fraud points. This suggests that fraudulent transactions have more variability in their feature space, which could be a useful insight for building classification models.

```python
# checking for multicollinearity with heatmap
plt.figure(figsize=(16,12))
corr = df.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={"shrink": .8})
plt.title('Feature Correlation Heatmap')
plt.savefig('../assets/corr.png', dpi=150, bbox_inches='tight')
plt.show()
```

There appears to be no significant multicollinearity among the features, as most correlation values are low. This is expected since the features V1-V28 are results of PCA transformation, which aims to produce uncorrelated components.
